```python
df1=pd.read_csv('用户收视信息.csv',encoding='gbk')
df2=pd.read_csv('用户回看信息.csv',encoding='gbk')
df3=pd.read_csv('用户点播信息.csv',encoding='gbk')
df4=pd.read_csv('用户单片点播信息.csv',encoding='gbk')
df=pd.DataFrame(np.zeros((1329,7)),columns=['凌晨','早晨','上午','中午','下午','傍晚','晚上'])
```

```python
def getTimeRange(s1,s2):
    res=[]
    stime=0;etime=0
    tlist=['凌晨','早晨','上午','中午','下午','傍晚','晚上']
    slist=[0,5,8,11,14,17,20,24]
    t1=float(s1.split(':')[0])+float(s1.split(':')[1])/60
    t2=float(s2.split(':')[0])+float(s2.split(':')[1])/60
    for s in range(len(slist)-1):
        if t1>=slist[s] and t1<slist[s+1]:
            stime=s
            break
    for s in range(len(slist)-1):
        if t2>slist[s] and t2<=slist[s+1]:
            etime=s
    if stime<=etime:
        res=tlist[stime:etime+1]
    else:
        res=tlist[:etime+1]+tlist[stime:]
    return res
```

```python
for i in range(len(df1)):
    print(i)
    tr=getTimeRange(df1.loc[i,'收看开始时间'],df1.loc[i,'收看结束时间'])
    for t in tr:
        df.loc[df1.loc[i,'机顶盒设备号']-10001,t]+=1
```

```python
for i in range(len(df2)):
    print(i)
    tr=getTimeRange(df2.loc[i,'回看开始时间'],df2.loc[i,'回看结束时间'])
    for t in tr:
        df.loc[df2.loc[i,'设备号']-10001,t]+=1
```

```python
for i in range(len(df4)):
    print(i)
    tr=getTimeRange(df4.loc[i,'观看开始时间'],df4.loc[i,'观看结束时间'])
    for t in tr:
        df.loc[df4.loc[i,'设备号']-10001,t]+=1
```

```python
#输出结果
res=pd.DataFrame(columns=['id','tag1','tag2','tag3'])
n=0
clm=df.columns.tolist()
for i in range(len(df)):
    for c in clm:
        if df.loc[i,c]>0.1:
            res.loc[n,'id']=i+10001
            res.loc[n,'tag1']="基本特征"
            res.loc[n,'tag2']="观看时间段"
            res.loc[n,'tag3']=c
            n+=1
```
