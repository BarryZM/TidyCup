```python
ds=pd.read_csv("电视产品信息数据.csv",encoding='gbk')
ds=ds.loc[ds.loc[:,'批次']=='少儿'].reset_index(drop=True)
etlist=[]#儿童节目的列表
for i in range(len(ds)):
    st=ds.loc[i,'连续剧分类']
    #print(st)
    #st=re.sub('：.*','',st)
    st=re.sub('\(\S+\)','',st)
    st=re.sub('【\S+】','',st)
    st=re.sub('（\S+）','',st)
    st=re.sub('\S+月\S+日','',st)
    st=re.sub('第.季','',st)
    st=re.sub('Ⅱ','',st)
    st=re.sub('Ⅲ','',st)
    #st=re.sub(' ','',st)
    st=st.strip()
    etlist.append(st)
etlist=sorted(list(set(etlist)))
```

```python
#增加儿童标签
for i in range(len(ra)):
    if ra.loc[i,'name'] in etlist and "儿童" not in json.loads(ra.loc[i,'info']):
        tagl=json.loads(ra.loc[i,'info'])
        if len(tagl)==0:
            tagl.append("动画")
        tagl.append("儿童")
        ra.loc[i,'info']=json.dumps(tagl)
        
```

```python
for i in range(len(ra)):
    ra.loc[i,'name']=ra.loc[i,'name'].split()[0]
    ra.loc[i,'name']=re.sub('\(\S+','',ra.loc[i,'name'])
```

```python
ra.to_csv('info2.csv',index=False)
```

```python
def count_time(s1,s2):
    t1=int(s1.split(':')[0])*60*60+int(s1.split(':')[1])*60+int(s1.split(':')[2])
    t2=int(s2.split(':')[0])*60*60+int(s2.split(':')[1])*60+int(s2.split(':')[2])
    if t2<t1:
        t2+=24*60*60
    return t2-t1
```
