```python
def count_time(s1,s2):
    t1=int(s1.split(':')[0])*60*60+int(s1.split(':')[1])*60+int(s1.split(':')[2])
    t2=int(s2.split(':')[0])*60*60+int(s2.split(':')[1])*60+int(s2.split(':')[2])
    if t2<t1:
        t2+=24*60*60
    return t2-t1
```

```python
typelist=['犯罪', '喜剧', '舞台艺术', '新闻', '儿童', '真人秀', '歌舞', '戏曲', '音乐', '冒险', '灾难', '恐怖', '家庭', '短片', '历史', '运动', '动画', '纪录片', '惊悚', '科幻', '战争', '动作', '健康', '脱口秀', '悬疑', '剧情', '古装', '鬼怪', '奇幻', '武侠', '同性', '体育', '西部', '传记', '爱情']
typelist=sorted(typelist)
```

```python
ra=pd.read_csv('info3.csv',encoding='gbk')
```

```python
#节目的标签矩阵
df=pd.DataFrame(np.zeros((len(ra),len(typelist))),columns=typelist)
for i in range(len(ra)):
    for t in json.loads(ra.loc[i,'info']):
        if t in typelist:
            df.loc[i,t]=1
```

```python
#节目的缩写
dfr=pd.read_csv("guankan.csv",encoding='gbk')
dfr.loc[:,'sname']=''
for i in range(len(dfr)):
    st=dfr.loc[i,'name']
    #print(st)
    st=re.sub('：.*','',st)
    st=re.sub('\(\S+\)','',st)
    st=re.sub('【\S+】','',st)
    st=re.sub('（\S+）','',st)
    st=re.sub('\S+月\S+日','',st)
    st=re.sub('第.季','',st)
    st=re.sub('Ⅱ','',st)
    st=re.sub('Ⅲ','',st)
    st=re.sub('\d{8}','',st)
    #st=re.sub(' ','',st)
    st=st.strip()
    dfr.loc[i,'sname']=st
```

```python
#缩写的dict，通过hash查找
dic={}
for i in range(len(ra)):
    dic[ra.loc[i,'name']]=json.loads(ra.loc[i,'info'])
```

```python
dfk=pd.DataFrame(np.zeros((1329,len(typelist))),columns=typelist)
for i in range(len(dfr)):
    #print(i)
    if dfr.loc[i,'sname'] not in dic.keys():
        print(i,dfr.loc[i,'sname'],dfr.loc[i,'name'])
        st=dfr.loc[i,'name']
        #print(st)
        st=re.sub('\(\S+\)','',st)
        st=re.sub('【\S+】','',st)
        st=re.sub('（\S+）','',st)
        st=re.sub('\S+月\S+日','',st)
        st=re.sub('第.季','',st)
        st=re.sub('Ⅱ','',st)
        st=re.sub('Ⅲ','',st)
        #st=re.sub(' ','',st)
        st=st.strip()
        taglist=dic[st]
    else:
        taglist=dic[dfr.loc[i,'sname']]
    if len(taglist)==0:
        continue
    timer=count_time(dfr.loc[i,'stime'],dfr.loc[i,'etime'])/len(taglist)
    for t in taglist:
        dfk.loc[dfr.loc[i,'id']-10001,t]+=timer
```
