```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('回看电视台.csv')
```

```python
def count_time(s1,s2):
    t1=int(s1.split(':')[0])*60*60+int(s1.split(':')[1])*60+int(s1.split(':')[2])
    t2=int(s2.split(':')[0])*60*60+int(s2.split(':')[1])*60+int(s2.split(':')[2])
    if t2<t1:
        t2+=24*60*60
    return t2-t1
```

```python
count_time(df.loc[0,'回看开始时间'],df.loc[0,'回看结束时间'])
```
