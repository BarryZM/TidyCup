```python
import pandas as pd
import numpy as np
import requests
import re
from bs4 import BeautifulSoup as BS
```

```python
df=pd.read_csv('点播信息全.csv',encoding='gbk')
```

```python
df=df.loc[:38009]
```

```python
df
```

```python
mlist=[]
for i in range(len(df)):
    st=df.loc[i,'影片名称']
    st=re.sub('：.*','',st)
    st=re.sub('\(\S+\)','',st)
    st=re.sub('【\S+】','',st)
    st=re.sub('\S+月\S+日','',st)
    st=re.sub('第.季','',st)
    st=re.sub('Ⅱ','',st)
    st=re.sub('Ⅲ','',st)
    st=re.sub(' ','',st)
    st=re.sub(' ','',st)
    
    st=st.strip()
    mlist.append(st)
```

```python
tvlist=sorted(list(set(mlist)))
len(tvlist)
```

```python
tvlist
```

```python
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.90 Safari/537.36"}
clist=['中文名','外文名','其它译名','出品公司', '制片地区', '拍摄地点', '发行公司', '首播时间', '导演', '编剧', '主演', '集数',
       '每集长度', '类型', '上映时间', '在线播放平台', '首播平台', '主题曲', '播出时间']
dfi=pd.DataFrame()
piclist=[]
for i in range(len(tvlist)):#len(tvlist)
    print(i)
    dfi.loc[i,'name']=tvlist[i]
    url="https://baike.baidu.com/item/"
    print(url+dfi.loc[i,'name'])
    try:
        res=requests.get(url+dfi.loc[i,'name'],headers=headers,timeout=10)
        res.encoding='utf-8'
    except Exception as e:
        print(e)
        continue
    soup=BS(res.text,'lxml')
    if soup.title.text=="百度百科——全球最大中文百科全书":
        print(dfi.loc[i,'oriname'])
    else:
        if soup.find('img',class_='picture'):
            piclist.append(soup.find('img',class_='picture')['src'])
            pic=requests.get(soup.find('img',class_='picture')['src'],headers=headers,timeout=10)
            with open('pic/'+tvlist[i]+'.jpg','wb') as f:
                f.write(pic.content)
        if soup.find('div',class_='lemma-summary'):
            dfi.loc[i,'info']=soup.find('div',class_='lemma-summary').text.strip()
        if soup.find('div',class_='basic-info'):
            dlist=soup.find('div',class_='basic-info').find_all('dt')
            for d in dlist:
                if d.text.replace('\xa0','') in clist:
                    dfi.loc[i,d.text.replace('\xa0','')]=d.next_sibling.next_sibling.text.strip()
```
